## `` Creating Standings, Game Statistics and Teams' forms Tables `` 

```python
def teamstats(team):
    scoredgoals = df[df.HomeTeam == team]["FTHomeGoal"].sum(axis=0)+df[df.AwayTeam == team]["FTAwayGoal"].sum(axis=0)
    concededgoals = df[df.HomeTeam == team]["FTAwayGoal"].sum(axis=0)+df[df.AwayTeam == team]["FTHomeGoal"].sum(axis=0)
    wins = df[(df.HomeTeam == team)&(df.FTResult == "H")].FTResult.count()+df[(df.AwayTeam == team)&(df.FTResult == "A")].FTResult.count()
    losses = df[(df.HomeTeam == team)&(df.FTResult == "A")].FTResult.count()+df[(df.AwayTeam == team)&(df.FTResult == "H")].FTResult.count()
    draws = df[(df.HomeTeam == team)&(df.FTResult == "D")].FTResult.count()+df[(df.AwayTeam == team)&(df.FTResult == "D")].FTResult.count()
    gd = scoredgoals - concededgoals
    points = 3*wins+draws
    cleansheets = df[(df.HomeTeam == team) & (df.FTAwayGoal == 0)].FTAwayGoal.count() + df[(df.AwayTeam == team) & (df.FTHomeGoal == 0)].FTHomeGoal.count()
    awaywins = df[(df.AwayTeam == team)&(df.FTResult == "A")].FTResult.count()
    homewins = df[(df.HomeTeam == team)&(df.FTResult == "H")].FTResult.count()
    roughindex = df[(df.HomeTeam == team)].HomeFoul.sum(axis=0)+df[(df.HomeTeam == team)].HomeYellow.sum(axis=0)+df[(df.HomeTeam == team)].HomeRed.sum(axis=0)+df[(df.AwayTeam == team)].AwayFoul.sum(axis=0)+df[(df.AwayTeam == team)].AwayYellow.sum(axis=0)+df[(df.AwayTeam == team)].AwayRed.sum(axis=0)
    outplayindex = df[(df.HomeTeam == team) & ((df.HomeShot+df.HomeCorner)>(df.AwayShot+df.AwayCorner))].HomeTeam.count() + df[(df.AwayTeam == team) & ((df.HomeShot+df.HomeCorner)<(df.AwayShot+df.AwayCorner))].AwayTeam.count()
    accuracy = round((df[(df.HomeTeam == team)].FTHomeGoal.sum(axis=0)+df[(df.AwayTeam == team)].FTAwayGoal.sum(axis=0))/(df[(df.HomeTeam == team)].HomeTarget.sum(axis=0)+df[(df.AwayTeam == team)].AwayTarget.sum(axis=0)), 2)
    comeback = df[(df.HomeTeam == team)&(df.HTResult == "A")&(df.FTResult == "H")].FTResult.count()+df[(df.AwayTeam == team)&(df.HTResult == "H")&(df.FTResult == "A")].FTResult.count()
    
    return [team, wins, draws, losses, scoredgoals, concededgoals, gd, points, cleansheets, awaywins, homewins, roughindex, outplayindex, accuracy, comeback]
```

```python
standings = {}
teams = []
wins = []
losses = []
draws = []
scored = []
conceded = []
gd = []
points = []
cleansheets = []
awaywins = []
homewins = []
winstreaks = []
roughindex = []
outplayindex = []
accuracy = []
comeback = []
morestats = {}
standings['Teams'] = teams
standings['Wins'] = wins
standings['Draws'] = draws
standings['Losses'] = losses
standings['ScoredGoals'] = scored
standings['ConcededGoals'] = conceded
standings['GD'] = gd
standings['Points'] = points
morestats['Teams'] = teams
morestats['CleanSheets'] = cleansheets
morestats['AwayWins'] = awaywins
morestats['HomeWins'] = homewins
morestats['WinStreak'] = winstreaks
morestats['RoughIndex'] = roughindex
morestats['OutplayIndex'] = outplayindex
morestats['Accuracy'] = accuracy
morestats['Comeback'] = comeback
```

```python
def winstreak(team):
    winstreak = 0
    winstreaklist = []
    for row, index in df[(df.HomeTeam == team) | (df.AwayTeam == team)].iterrows():
        if ((index[1] == team) & (index[5] == "H")) or ((index[2] == team) & (index[5] == "A")):
            winstreak += 1
        else:
            winstreaklist.append(winstreak)
            winstreak = 0
    return max(winstreaklist)
```

```python
for i in df["HomeTeam"].unique():
    a = teamstats(i)
    teams.append(a[0])
    wins.append(a[1])
    draws.append(a[2])
    losses.append(a[3])
    scored.append(a[4])
    conceded.append(a[5])
    gd.append(a[6])
    points.append(a[7])
    cleansheets.append(a[8])
    awaywins.append(a[9])
    homewins.append(a[10])
    roughindex.append(a[11])
    outplayindex.append(a[12])
    b = winstreak(i)
    winstreaks.append(b)
    accuracy.append(a[13])
    comeback.append(a[14])
```

## By exploring different aspects of the fixture tables, we were able to understand how the results of the season came about. 

### `` The top 4 teams Manchester City, Liverpool, Chelsea and Arsenal appeared consistently in the strong stats tables like scored goals, clean sheets, outplayindex, ... indicating a strong attacking and defending tactics compared to other teams in the league. Moreover, some of them (especially Man City and Liverpool) also appeared to have most wins, longest winning streaks as well as games with big goal margins, indicating a consistent and stable form throughout the season. That explains how these strong teams ended up at the top of the table.``

### `` When we look at the form graph, the top 5 teams had really stable form. Arsenal started well with more wins and less losses in the first half of the season; however, not being able to maintain such good streaks makes them fail to stay at the top of the table after a long time. Chelsea and Man City were more stable with their forms. Liverpool also achieved roughly the same number of wins, draws and losses compared to Chelsea and Man City; Liverpool's 11-game winning streak in the second half of the season made the race to the title much more exciting. However, their loss to Chelsea and draw to Crystal Palace at the very end of the season ruined their dream to win the title.``

### ``Manchester United, the previous season Champion, showed a fairly good playing statistics (in terms of scored goals and finishing accuracy); however, with the arrival of the new coach - David Moyes, they were not able to maintain their winning form and fell out of top 4.``

### ``Southampton appeared to be a team with an interesting tactics. Not starting with a lot of stars in the team, the team still managed to play in a fierce style; appearing in top teams with most clean sheets, outplaying the opponents and rough playstyle.``

### ``The three relegated teams - Cardiff City, Norwich and Fulham - usually appeared in teams with most conceded goals and most losses as well as possessed an unstable form with more losses and draws than wins.``
